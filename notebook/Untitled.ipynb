```python
import geopandas as gpd
import pandas as pd
import glob, os, requests
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.github.com/ssujit/mobility_explorer/main/data/'
wkam = 'wkamo_hdn.gpkg'
wkpm = 'wkpmo_hdn.gpkg'
satam = 'satamo_hdn.gpkg'
satpm = 'satpmo_hdn.gpkg'
sunam = 'sunamo_hdn.gpkg'
sunpm = 'sunamo_hdn.gpkg'

wk_am = gpd.read_file(f'{url}{wkam}')
wk_pm = gpd.read_file(f'{url}{wkpm}')
sat_am = gpd.read_file(f'{url}{satam}')
sat_pm = gpd.read_file(f'{url}{satpm}')
sun_am = gpd.read_file(f'{url}{sunam}')
sun_pm = gpd.read_file(f'{url}{sunpm}')
```

```python
wk_am.head()
```

```python
wk_am.plot("pti", cmap="RdYlGn_r", legend=True)
```

```python
# Set up subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 18))

# Plot each GeoDataFrame
wk_am.plot("pti", cmap="RdYlGn_r", legend=True, ax=axes[0, 0])
axes[0, 0].set_title("Weekday AM")

wk_pm.plot("pti", cmap="RdYlGn_r", legend=True, ax=axes[0, 1])
axes[0, 1].set_title("Weekday PM")

sat_am.plot("pti", cmap="RdYlGn_r", legend=True, ax=axes[1, 0])
axes[1, 0].set_title("Saturday AM")

sat_pm.plot("pti", cmap="RdYlGn_r", legend=True, ax=axes[1, 1])
axes[1, 1].set_title("Saturday PM")

sun_am.plot("pti", cmap="RdYlGn_r", legend=True, ax=axes[2, 0])
axes[2, 0].set_title("Sunday AM")

sun_pm.plot("pti", cmap="RdYlGn_r", legend=True, ax=axes[2, 1])
axes[2, 1].set_title("Sunday PM")

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.boxplot(wk_am['pti'].dropna(), vert =False)
# Add labels and title
plt.xlabel('values')
plt.ylabel('pti')
plt.title('Boxplot for pti in wk_am')

# Show the plot
plt.show()
```

```python
pti_values = [df['pti'].dropna() for df in [wk_am, wk_pm, sat_am, sat_pm, sun_am, sun_pm]]
plt.boxplot(pti_values, labels=['wk_am', 'wk_pm', 'sat_am', 'sat_pm', 'sun_am', 'sun_pm'], vert=False)

# Add labels and title
plt.xlabel('pti Values')
plt.ylabel('time')
plt.title('Boxplot for pti')

# Show the plot
plt.show()
```
